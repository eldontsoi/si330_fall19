# Question 1: Data Cleaning
Below is a list of average annual snowfall totals in the United States by state for the period mid-1985 to mid-2015.

## Part A
We need to transform this wikipedia dataset into a well-formated table with the following columns:
1. **State**: The state name. No digits should be present.
2. **Average_annual_snowfall_in_inches**.
3. **Elevation_in_feet**. Having commas in this column is fine.
4. **Elevation_in_m**. Having commas in this column is fine.

**Note**: Also, please sort the rows by **Average_annual_snowfall_in_inches** in ascending order (so the state with least snow should appear first.)

```python
import pandas as pd
import numpy as np
data = pd.read_html("https://en.wikipedia.org/wiki/List_of_snowiest_places_in_the_United_States_by_state")[0]
data.head()
```

```python
# Your code here
```

```python
# Solution code here
data['State'] = data['State'].str.replace(r'\d*\.','')
data['Average_annual_snowfall_in_inches'] = data['Average annual snowfall[3]'].str.extract('(?P<Average_annual_snowfall_in_inches>\d+\.*\d*)')
data['Elevation_in_feet'] = data['elevation[4]'].str.extract('(?P<Elevation_in_feet>\d+,*\d*)')
data['Elevation_in_m'] = data['elevation[4]'].str.extract('(?P<Average_annual_snowfall_in_cm>\(\d+,*\d*\.*\d*)')
data['Elevation_in_m'] = data['Elevation_in_m'].str.replace(r'\(','')

data = data[['State','Average_annual_snowfall_in_inches','Elevation_in_feet','Elevation_in_m']]
data['Average_annual_snowfall_in_inches'] = pd.to_numeric(data['Average_annual_snowfall_in_inches'])
data = data.sort_values(by=['Average_annual_snowfall_in_inches'])

data.head()
```

# Question 2: Data Manipulation
My roommate and I are trying to get housing for the next year. My main task is to find all the nearby apartments and contact information, and my roommate's job is to call each leasing office for apartment availability and pricing. As expected, some apartments do not have current vacancy.

```python
import pandas as pd
import numpy as np

my_df=pd.DataFrame([['#1',2,'Central'],
                   ['#2',1,'North'],
                   ['#3',2,'North'],
                   ['#4',1,'North'],
                   ['#5',2,'Central'],
                   ['#6',1,'Central'],
                   ['#7',1,'North'],
                   ['#8',1,'Central'],
                   ['#9',2,'North'],
                   ['#10',1,'Central']], columns=["name","number of bathrooms","location"])
roommate_df=pd.DataFrame([['#1',1350],
                         ['#2',750],
                         ['#3',820],
                         ['#4',802],
                         ['#6',1100],
                         ['#8',1210],
                         ['#9',900]], columns=["name","price per person"])

display(my_df)
display(roommate_df)
```

## Part A
Write a function to return the mean and standard deviations (from `numpy`) of the average price per person for central campus and north campus apartments. Only include apartments with vacancies.

```python
# Your code here
```

```python
# Solution code here
data=pd.merge(my_df,roommate_df,left_on='name',right_on='name', how="right")
data[['location','price per person']].groupby("location").agg([np.mean,np.std])
```
