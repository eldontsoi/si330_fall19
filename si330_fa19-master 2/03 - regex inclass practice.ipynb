```python
data=''
with open('hockey.txt','r') as f:
    #print(len(f.readlines()))
    data="".join(f.readlines())
data
```

```python
import re
re.findall(r'\d\d\d\d–\d\d: .*',data)
```

```python
for item in re.finditer(r"(\d\d\d\d)(?:–)(\d\d)(: )(.*)",data):
    print(item.groups())
#(?P<name>)
```
