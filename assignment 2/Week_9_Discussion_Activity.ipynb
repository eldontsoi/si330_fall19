### PageRank


1.  We give each node an amount of “status” (or PageRank). At each time step the node distributes that PageRank to other nodes.

2. At each time step, we take a certain portion of PageRank from all nodes and redistribute it evenly. (The damping parameter, or fraction of PageRank that does not get redistributed, was called d in the lecture, s in the book, and will be called  in NetworkX).

Node A’s PageRank is the amount of status that it has when this process stabilizes.



To compute Pagerank, you should know about 3 concepts and how they are generated using networkx:
1. Directed Graph - It consists of a set of nodes, with a set of directed edges; each directed edge is a link from one node to another, with the direction being important.* {Syntax: g2 = nx.DiGraph()}*
2. In-degree- It is the number of links coming into a node *{Syntax:G.in_degree('A')}*
3. Out-degree- It is the number of links coming out a node *{Syntax:G.out_degree('A')}*

```
import networkx as nx
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```
g2 = nx.DiGraph()
g2_edges = [("A","B"),("A","C"),("B","C"),("C","D"),("D","A")]
g2.add_edges_from(g2_edges)
nx.draw_networkx(g2, with_labels=True)
plt.show()
```

```
print('In-degree for Node A:', g2.in_degree('A'))
print('Out-degree for Node A:', g2.out_degree('A'))
```
